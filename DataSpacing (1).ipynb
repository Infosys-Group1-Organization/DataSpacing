```python
import pandas as pd

df = pd.read_csv('/content/Spacing.csv')

print(df.to_string()) 
```

```python
def spacetrim(dataframe):
   
    for i in dataframe.columns:
         
        if dataframe[i].dtype == 'object':
             
            dataframe[i] = dataframe[i].map(str.strip)
        else:
            pass
 
spacetrim(df)
print(df)
```
